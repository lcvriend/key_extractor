```python
import pandas as pd
import key_extractor
from random import randint, choice
```

```python
n = 1000
df = pd.DataFrame(index=range(n))
df['col1'] = [randint(10000, 70000) for _ in range(n)]
df['col2'] = [choice(['x', 'y']) for _ in range(n)]
df['col3'] = [choice(['A', 'B', 'C', 'D']) for _ in range(n)]
df.askeys('col1', batch_size=100, to='print', groupby=['col2', 'col3'])
```

```python
df.askeys.to_file('.', 'col1', batch_size=500, groupby='col2')
```
